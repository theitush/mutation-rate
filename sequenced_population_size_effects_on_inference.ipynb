```python
import numpy as np
import matplotlib.pyplot as plt
from modeling import simulate_data, run_methods
```

# sequence_sample_size determines the amount of noise:

```python
generations_number = 20
wt_freqs = 0.0001 * np.ones(10)
fitness=1.2
mutation_rate = 10**-5
population_size = 10**9

simulate_data(generations_number=generations_number, wt_freqs=wt_freqs, population_size=population_size, 
              fitness=fitness, mutation_rate=mutation_rate, 
              sequence_sample_size=10**4, color='black', label="sequenced sample size = 10^4", plot=True)
simulate_data(generations_number=generations_number, wt_freqs=wt_freqs, population_size=population_size, 
              fitness=fitness, mutation_rate=mutation_rate, 
              sequence_sample_size=10**7, color='orange', label="sequenced sample size = 10^7", plot=True)
plt.show()
```

# <u> Fitness Inference

## With low noise, inferring fitness is easy:

```python
generations_number = 10
wt_freqs = 0 * np.ones(10)
mutation_rate = 10**-5
population_size = 10**9
```

```python
sequence_sample_size = 10**5
```

```python
simulate_data(generations_number=generations_number, wt_freqs=wt_freqs, population_size=population_size, 
              fitness=1, mutation_rate=mutation_rate, 
              sequence_sample_size=sequence_sample_size, color='blue', plot=True)
simulate_data(generations_number=generations_number, wt_freqs=wt_freqs, population_size=population_size, 
              fitness=1.5, mutation_rate=mutation_rate, 
              sequence_sample_size=sequence_sample_size, color='orange', plot=True)
plt.title('Fitness Inference: Low Noise, Low Intial Freq')
plt.show()
```

## With noise, it can be harder:

```python
sequence_sample_size = 10**2
```

```python
simulate_data(generations_number=generations_number, wt_freqs=wt_freqs, population_size=population_size, 
              fitness=1, mutation_rate=mutation_rate, 
              sequence_sample_size=sequence_sample_size, color='blue', plot=True)
simulate_data(generations_number=generations_number, wt_freqs=wt_freqs, population_size=population_size, 
              fitness=1.5, mutation_rate=mutation_rate, 
              sequence_sample_size=sequence_sample_size, color='orange', plot=True)
plt.title('Fitness Inference: High Noise, Low Intial Freq')

plt.show()
```

## But can still work if the initial freq is high enough:

```python
wt_freqs = 0.01 * np.ones(10)
```

```python
simulate_data(generations_number=generations_number, wt_freqs=wt_freqs, population_size=population_size, 
              fitness=1, mutation_rate=mutation_rate, 
              sequence_sample_size=sequence_sample_size, color='blue', plot=True)
simulate_data(generations_number=generations_number, wt_freqs=wt_freqs, population_size=population_size, 
              fitness=1.5, mutation_rate=mutation_rate, 
              sequence_sample_size=sequence_sample_size, color='orange', plot=True)
plt.title('Fitness Inference: High Noise, High Intial Freq')
plt.show()
```

# <u> Mutation rate inference

##  also easy when noise is low:

```python
wt_freqs = np.zeros(10)
fitness = 1
```

```python
sequence_sample_size = 10**5
```

```python
simulate_data(generations_number=generations_number, wt_freqs=wt_freqs, population_size=population_size, 
              fitness=fitness, mutation_rate=10**-6, 
              sequence_sample_size=sequence_sample_size, color='blue', plot=True)
simulate_data(generations_number=generations_number, wt_freqs=wt_freqs, population_size=population_size, 
              fitness=fitness, mutation_rate=10**-5, 
              sequence_sample_size=sequence_sample_size, color='orange', plot=True)
plt.title('Mutation Rate Inference: Low Noise')
plt.show()
```

## And get's harder with more noise:

```python
sequence_sample_size = 10**3
```

```python
simulate_data(generations_number=generations_number, wt_freqs=wt_freqs, population_size=population_size, 
              fitness=fitness, mutation_rate=10**-6, 
              sequence_sample_size=sequence_sample_size, color='blue', plot=True)
simulate_data(generations_number=generations_number, wt_freqs=wt_freqs, population_size=population_size, 
              fitness=fitness, mutation_rate=10**-5, 
              sequence_sample_size=sequence_sample_size, color='orange', plot=True)
plt.title('Mutation Rate Inference: High Noise')
plt.show()
```

## Specifically, looks like sequence_sample_size > 1 / mutation_rate is good enough and maybe we can even work with slightly lower values..

```python
sequence_sample_size = 10**4
```

```python
simulate_data(generations_number=generations_number, wt_freqs=wt_freqs, population_size=population_size, 
              fitness=fitness, mutation_rate=10**-6, 
              sequence_sample_size=sequence_sample_size, color='blue', plot=True)
simulate_data(generations_number=generations_number, wt_freqs=wt_freqs, population_size=population_size, 
              fitness=fitness, mutation_rate=10**-5, 
              sequence_sample_size=sequence_sample_size, color='orange', plot=True)
plt.title('Mutation Rate Inference: Medium Noise')
plt.show()
```

We should also remember that higher intial frequencies will make everything harder as mutation rate becomes neglible pretty fast.... but that is a topic for a [different notebook...](https://github.com/theitush/mutation-rate/blob/master/hetrogenic_initial_frequency.ipynb)
