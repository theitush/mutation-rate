```python
def create_pivot_table(df):
    pivot_table = pd.DataFrame(df.groupby(['ref_base']).read_base.value_counts()).rename(
        columns={'read_base': 'count'}).reset_index()
    pivot_table = pivot_table.pivot(index='ref_base', columns='read_base', values='count')
    return pivot_table
```

```python
def reg_plot(x, y, size=None, save_to=None):
    if size is None:
        size = 10
    max_val = max(max(x), max(y))
    min_val = min(min(x), min(y))
    set_plots_size_params(size*2)
    plt.figure(figsize=(size*1.5, size))
    plt.scatter(x, y)
    plt.plot([min_val, max_val], [min_val, max_val])
    try:
        plt.xlabel(x.name)
        plt.ylabel(y.name)
    except:
        print("Couldn't set axes names. Trying putting pd.Series objects as x & y")
    if save_to:
        plt.savefig(save_to)
```

```python
from scipy.stats import ks_2samp
```

```python
def compare_dists(df):
    loop = df[df.source=='loop']
    illumina = df[df.source=='illumina']
    if loop.empty:
        return
    return ks_2samp(loop.frequency, illumina.frequency)
```

```python
top_mutations = [f'{pos}{base}' for (pos, base) in top_mutations]
color_list = ['#e6194B', '#3cb44b', '#ffe119', '#4363d8', '#f58231', '#911eb4', '#42d4f4', '#f032e6', '#bfef45',
              '#fabed4', '#469990', '#dcbeff', '#9A6324', '#fffac8', '#800000', '#aaffc3', '#808000', '#ffd8b1',
              '#000075', '#000000']
color_dict = {top_mutations[i]: color_list[i] for i in range(len(top_mutations))}
```

```python
def mutation_time_series(df, ax, colors_dict, going=None, just_one=None):
    if just_one:
        if df.source.iloc[0] != just_one:
            return
    x = df.passage
    y = df.Freq
    if going=='up':
        if y.iloc[0] >= y.iloc[-1]*0.7:
            return
    if going=='down':
        if y.iloc[0] <= y.iloc[-1]*0.7:
            return
    label = f"{df.Pos.iloc[0]}{df.Base.iloc[0]}"
    if label in colors_dict.keys():
        color = colors_dict[label]
        alpha = 1
    else:
        color = 'black'
        label = None
        alpha = 0.1
    if df.source.iloc[0] == "loop":
        linestyle='dashed'
    else:
        linestyle='solid'
    ax.plot(df.passage, df.Freq, color=color, label=label, alpha=alpha, linestyle=linestyle)
    # ax.text(x.iloc[-1], y.iloc[-1], label)  # <- text next to plot!
    return ax

```

```python
def plot_compare_loop_illumina(mutations):
    fig, axes = plt.subplots(figsize=(60, 30), ncols=3, nrows=3)
    plt.suptitle("MS2 Time Series MOI 0.1 with Read_count>50", fontsize=32)
    plt.subplots_adjust(wspace=0.4, hspace=0.3)
    y_max = mutations.Freq.max()*1.1
    
    for i in range(3):
        title = f"Mutations with Frequency < {round(mutations.Freq.max(),2)} in Sample {i+1}"
        km_muts = mutations[mutations.km==i+1]
        km_muts.groupby(['source','Pos', 'Base']).apply(
            lambda df: mutation_time_series(df, axes[i][2], color_dict))
        axes[i][2].set_ylim([-0.001, y_max])
        axes[i][2].legend(bbox_to_anchor=(1, 1))
        axes[i][2].set_title(f"{title} (combined)")
        axes[i][2].set_xlabel('Passage')
        axes[i][2].set_ylabel('Frequency')
        km_muts.groupby(['source','Pos', 'Base']).apply(
            lambda df: mutation_time_series(df, axes[i][1], color_dict, just_one='loop'))
        axes[i][1].set_ylim([-0.001, y_max])
        axes[i][1].legend(bbox_to_anchor=(1, 1))
        axes[i][1].set_title(f"{title} (loop data)")
        axes[i][1].set_xlabel('Passage')
        axes[i][1].set_ylabel('Frequency')
        km_muts.groupby(['source','Pos', 'Base']).apply(
            lambda df: mutation_time_series(df, axes[i][0], color_dict, just_one='illumina'))
        axes[i][0].set_ylim([-0.001, y_max])
        axes[i][0].legend(bbox_to_anchor=(1, 1))
        axes[i][0].set_title(f"{title} (illumina data)")
        axes[i][0].set_xlabel('Passage')
        axes[i][0].set_ylabel('Frequency')
    return fig, axes
```

```python
fig, axes = plot_compare_loop_illumina(mutations)
```
