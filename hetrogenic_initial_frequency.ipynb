# 1. Intial frequency & inference

## low intial frequency --> fitness is neglible 
## high intial frequency --> mutation rate is neglible

```python
gens = 17
ps = 10 ** 8
ss = 10 ** 5
freq = 0.005
simulate_data(generations_number=gens, wt_freqs=freq*np.ones(10), population_size=ps, fitness=1.15, 
              mutation_rate=10**-6, 
              sequence_sample_size=ss, color='black', label=None, plot=True)
simulate_data(generations_number=gens, wt_freqs=freq*np.ones(10), population_size=ps, fitness=1.05, 
              mutation_rate=10**-3, 
              sequence_sample_size=ss, color='blue', label=None, plot=True)
plt.show()
```

## Let's see what that means for inference:

### low frequency --> accurate mutation rate, inaccurate fitness:

```python
dfs = run_methods(10**-5, 0.8, 0.0001, wt_freqs=0*np.ones(10), methods=['avg'])
```

### high frequency --> inaccurate mutation rate, accurate fitness:

```python
dfs = run_methods(10**-5, 0.8, 0.0001, wt_freqs=0.1*np.ones(10), methods=['avg'])
```

# 2. Inference of multiple alleles with different intial frequency

## All our methods give the same results when the intial frequencies are the same:

```python
dfs = run_methods(10**-5, 0.8, 0.0001, wt_freqs=0.1*np.ones(3), methods='All')
```

```python
dfs = run_methods(10**-5, 0.8, 0.0001, wt_freqs=0*np.ones(3), methods='All')
```

## It starts getting interesting when we use hetrogenic intial frequencies:

### With a big diff between freqs, it's better to concat the posteriors

```python
dfs = run_methods(10**-5, 0.8, 0.0001, wt_freqs=[0, 0.1], methods=['megapost', 'megadist'])
```

### but if the diff is small enough, inferring everything at once gives us the best of both worlds:

```python
dfs = run_methods(10**-5, 0.8, 0.0001, wt_freqs=[0, 0.0001], methods=['megapost', 'megadist'])
```
